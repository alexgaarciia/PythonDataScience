# Learning Matplotlib
It is a graph plotting library in python that serves as a visualization utility.

# Installation of Matplotlib

```python
import matplotlib
```

## Pyplot
Most of the Matplotlib utilities lies under the `pyplot` submodule.

```python
import matplotlib.pyplot as plt
```

# Plotting
The `plot()` function is used to draw points in a diagram. By default, it draws a line from point to point.

It takes some parameters:
- Parameter 1: Array containing the points on the x-axis.
- Parameter 2: Array containing the points on the y-axis.

If we need to plot a line from (1,3) to (8,10), we have to pass two arrays [1,8] and [3, 10] to the plot function.

```python
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## Plotting without line
To plot only the markers, we can use the `shortcut strin notation` parameter 'o', which means 'rings'.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

## Multiple points
You can plot as many points as you like, just **make sure you have the same number of points in both axis**.

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

# Markers

The keyword argument `marker` is used to emphasize each point with a specified marker.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'x')
plt.show()
```

## Format Strings `fmt`
You can also use the `shortcut string notation` parameter to specify the marker. This is called `fmt`and is written with this syntax: **`marker|line|color`** (tabs not included).

```python
ypoint = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()

plt.plot(ypoints, 'o-.b')
plt.show()
```

## Line reference
Line reference:
| Line Syntax | Description         |
|------------ | ------------------- |
| '-'         | Solid line          |
| ':'         | Dotted line         |
| '--'        | Dashed line         |
| '-.'        | Dashed/dotted line  |

## Marker size
`ms` allows us to se the size of the markers.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

## Marker color
`mec` (marker edge color) sets the color of the edge of the markers.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

We can also use the keyword `mfc` (marker face color) to set the color inside the edge of the markers.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r', mec = 'r')
plt.show()
```
