# Learning Matplotlib
It is a graph plotting library in python that serves as a visualization utility.

# Installation of Matplotlib

```python
import matplotlib
```

## Pyplot
Most of the Matplotlib utilities lies under the `pyplot` submodule.

```python
import matplotlib.pyplot as plt
```

# Plotting
The `plot()` function is used to draw points in a diagram. By default, it draws a line from point to point.

It takes some parameters:
- Parameter 1: Array containing the points on the x-axis.
- Parameter 2: Array containing the points on the y-axis.

If we need to plot a line from (1,3) to (8,10), we have to pass two arrays [1,8] and [3, 10] to the plot function.

```python
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## Plotting without line
To plot only the markers, we can use the `shortcut strin notation` parameter 'o', which means 'rings'.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

## Multiple points
You can plot as many points as you like, just **make sure you have the same number of points in both axis**.

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

# Markers

The keyword argument `marker` is used to emphasize each point with a specified marker.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'x')
plt.show()
```

## Format Strings `fmt`
You can also use the `shortcut string notation` parameter to specify the marker. This is called `fmt`and is written with this syntax: **`marker|line|color`** (tabs not included).

```python
ypoint = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()

plt.plot(ypoints, 'o-.b')
plt.show()
```

## Line reference
Line reference:
| Line Syntax | Description         |
|------------ | ------------------- |
| '-'         | Solid line          |
| ':'         | Dotted line         |
| '--'        | Dashed line         |
| '-.'        | Dashed/dotted line  |

## Marker size
`ms` allows us to se the size of the markers.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

## Marker color
`mec` (marker edge color) sets the color of the edge of the markers.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

We can also use the keyword `mfc` (marker face color) to set the color inside the edge of the markers.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r', mec = 'r')
plt.show()
```

# Line
We can use the keyword argument `linestyle` or `ls` to change the style of the plotted line.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls="dotted")
plt.show()

plt.plot(ypoints, ls="dashed")
plt.show()
```

## Line color
We can use the argument `color` or `c` to set the color of the line.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls="dotted", color = 'r')
plt.show()
```

## Line width
We can use the keyword `linewidth` or `lw` to change the width of the line.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

## Multiple lines

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

# Multiple labels and title
We use `xlabel()` and `ylabel()` to set a label for the x and y axis.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

## Create a title
This is done using the `title()` function.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

## Set font properties for title and labels
We can use the `fontdict()` parameter in **xlabel(), ylabel(), and title()** to set font properties for the title and labels.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {"family":"serif", "color":"blue", "size":20}
font2 = {"family":"serif", "color":"darkred", "size":15}


plt.plot(x, y)
plt.title("Sports Watch Data", fontdict=font1)
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)
plt.show()
```

## Position the title
This is adjusted using the `loc` parameter.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

# Subplot
With the `subplot()` function we can draw multiple plots in one figure. It takes **three** arguments that describe the layout of the figure:
- First: Number of rows.
- Second: Number of columns.
- Third: Index of the current plot.

```python
# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)


# Plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

Let's draw two plots on top of each other:

```python
# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x, y)


# Plot 2: 
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x, y)
plt.show()
```

We can even add a title to each plot:

```python
# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("SALES")


# Plot 2: 
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("INCOME")
plt.show()
```

## Super title
This is used to add a title to the entire figure using `suptitle()`.

```python
# Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("SALES")


# Plot 2: 
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

# Scatter 
These are created using the function `scatter()`.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

It is possible to compare plots as we did with the lines:

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)


#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x, y)
plt.show()
```

## Colors
As in lines, these can be adjusted using the argument `color` or `c`.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, c = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, c = '#88c999')

plt.show() 
```

## Color each dot
This can be done by specifying an array of colors as value for the `c` argument.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)
plt.show()
```

## ColorMap
A colormap is a list of coors, where each color has a value that ranges from 0 to 100. 

This can be specified with the keyword argument `cmap` with the value of the colormap (such as `virids`). 

In addition you have to create an array with values (from 0 to 100), one value for each point in the scatter plot.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap="viridis")
plt.colorbar()
plt.show()
```

## Size
The size of the dots can be adjusted with the `s` argument. 

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)
plt.show() 
```

## Alpha (to adjust the transparency of the dots)

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)
plt.show() 
```

# Bars
Created using the `bar()` function.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

## Horizontal bars (`barh()`)

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

## Color
Set using the `color` argument.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

## Width and height
Set using the parameters `width` (for vertical bars) and `height` (for horizontal bars).

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

# Pie charts (`pie()`)

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

## Labels
Add labels to pie chart with the `labels` parameter:

```python
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels)
plt.show()
```

## Start angle
The start angle can be changed with the parameter `startangle`.

```python
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 

plt.pie(y, labels = mylabels, startangle = 180)
plt.show() 

plt.pie(y, labels = mylabels, startangle = 270)
plt.show() 

plt.pie(y, labels = mylabels, startangle = 0)
plt.show() 
```

## Explode
If we want to stand out one of the wedges, the `explode` parameter allows us to do that. 

The explode parameter must be an array with one value for each wedge. Each value represents how far from the center each wedge is displayed.

```python
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

## Shadow
A shadow can be added to the pie chard by setting the `shadows` parameter to `True`.

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

## Colors
We can set the color of each wedge with the `colors` parameter. If specified, it must be an array with one value for each wedge.

```python
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

## Legend

```python
plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

We can even add a header to the legend by adding the `title` parameter to the `legend` function:

```python
plt.pie(y, labels = mylabels)
plt.legend(title="Four Fruits:")
plt.show() 
```
